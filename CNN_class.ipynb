```python
import numpy as np
import h5py
#import h5
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hf = h5py.File('./Noise_Data/sigma10_classes/one.h5', 'r')
hf.keys()
one_data=hf.get('dataset')

one_data.shape
```

```python
plt.figure(figsize=(16, 16))
fig, ax = plt.subplots(2, 2)
count=0
for i in range(2):
    for j in range(2):
         data = np.array([[i, j], [i+0.5, j+0.5]])
         im = ax[i, j].imshow(one_data[count,:,1])
         plt.colorbar(im, ax=ax[i, j])
         count+=1
plt.setp(ax[-1, :], xlabel='x-axis[Mpc]')
plt.setp(ax[:, 0], ylabel='y-axis[Mpc]')
plt.show()
```
